```python
import pandas as pd
```

```python
d1 = pd.read_csv('origin data\\kolesa_cars1.csv')
d2 = pd.read_csv('origin data\\kolesa_cars2.csv')
d3 = pd.read_csv('origin data\\kolesa_cars3.csv')
```

```python
splits1 = d1['title'].str.split(' ', expand=True)
splits2 = d2['title'].str.split(' ', expand=True)
splits3 = d3['title'].str.split(' ', expand=True)
```

```python
d1['brand'] = splits1[0]
d2['brand'] = splits2[0]
d3['brand'] = splits3[0]

d1['model'] = splits1.iloc[:, 2:].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
d2['model'] = splits2.iloc[:, 1:].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
d3['model'] = splits3.iloc[:, 1:].apply(lambda x: ' '.join(x.dropna().astype(str)), axis=1)
```

```python
d1 = d1.drop(columns=['title'])
d2 = d2.drop(columns=['title'])
d3 = d3.drop(columns=['title'])
```

```python
df = pd.concat([d1, d2, d3], ignore_index=True)
```

```python
df[50:70]
```

```python
df.info()
```

```python
df['year'].unique()
```

```python
df = df[df['year'] != 'На заказ']
```

```python
df['year'] = pd.to_numeric(df['year'])
```

```python
df['volume'].unique()
```

```python
df = df[df['volume'] != 'эектричество']
df['volume'] = pd.to_numeric(df['volume'])
```

```python
df['price'] = df['price'].astype(str).str.replace(r'\D+', '', regex=True)
df['price'] = pd.to_numeric(df['price'])
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv('kolesa_cars.csv', index=False, encoding='utf-8-sig')
```
